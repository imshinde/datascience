### Sql Database Table

```python
import sqlite3

con = sqlite3.connect('my_data.db')
cur = con.cursor()
cur.execute("""CREATE TABLE IF NOT EXISTS my_inventory
            (prod_id VARCHAR(100) NOT NULL,
            prod_name VARCHAR(250) NOT NULL,
            prod_price DECIMAL(10,2) NOT NULL,
            prod_quant INT NOT NULL,
            PRIMARY KEY(prod_id))""")
cur.close()
con.close()
```

### Inventory Management System

```python
import tkinter as tk
from tkinter import *
from tkinter import ttk, messagebox
import sqlite3

root = tk.Tk()
root.title('Mini Inventory Management System')

tab_control = ttk.Notebook(root)
my_frame = ttk.Frame(tab_control)
lblframe = ttk.LabelFrame(my_frame,text="Inventory Management")
lblframe1 = ttk.LabelFrame(my_frame, text='Product List', borderwidth=3)

lblframe.grid(row=0,column=0,padx=8,pady=4,sticky="N")
lblframe1.grid(row=0,column=1,padx=8,pady=4,sticky="N")


tree = ttk.Treeview(lblframe1, columns=('Product_ID','Product_Name','Selling_Price','Quantity'), height=20)
tree.place(x=30,y=95)
vsb = ttk.Scrollbar(tree, orient="vertical",command=tree.yview)
vsb.place(x=30+954+5,y=0,height=200+220)

tree.configure(yscrollcommand=vsb.set)
tree.heading('#0',text='Item No')
tree.heading('#1',text='Product ID')
tree.heading('#2',text='Product Name')
tree.heading('#3',text='Selling Price')
tree.heading('#4',text='Quantity')
tree.column('#0',stretch=tk.YES)
tree.column('#1',stretch=tk.YES)
tree.column('#2',stretch=tk.YES)
tree.column('#3',stretch=tk.YES)
tree.column('#4',stretch=tk.YES)
tree.grid(row=10, columnspan=4,sticky='nsew')


def insert_info():
    try:
        if PRO_ID.get() !="" and PRO_NAME.get() !="" and PRO_PRICE.get()!="" and PRO_QUANT.get()!="":
            con = sqlite3.connect('my_data.db')
            cur = con.cursor()
            cur.execute('insert into my_inventory(prod_id,prod_name,prod_price,prod_quant) values(?,?,?,?)',
                        [PRO_ID.get(),str(PRO_NAME.get()).lower(),PRO_PRICE.get(),PRO_QUANT.get()])
            con.commit()
            messagebox.showinfo("","inserted successfully")
        else:
            messagebox.showinfo("","fill all fields.")
    
    except tk.TclError:
        messagebox.showinfo("","price and quantity should be a number.")
        
    except sqlite3.IntegrityError:
        messagebox.showwarning("","product id already exist.")

def show_info():
    try:
        i=0
        tree.delete(*tree.get_children())
        con = sqlite3.connect('my_data.db')
        cur = con.cursor()
        results = cur.execute('select * from my_inventory')
        for row in results:
            tree.insert('','end',text="item "+str(i),values=(row[0],row[1],row[2],row[3]))
            i=i+1
        cur.close()
    except:
        messagebox.showwarning("","no data found.")

global PRO_ID
PRO_ID = tk.StringVar()

global PRO_NAME
PRO_NAME = tk.StringVar()

global PRO_PRICE
PRO_PRICE = tk.DoubleVar()
PRO_PRICE.set("")

global PRO_QUANT
PRO_QUANT = tk.IntVar()
PRO_QUANT.set("")

pro_id = ttk.Label(lblframe, text="Product ID:")
id_entry = ttk.Entry(lblframe, textvariable=PRO_ID, width=25)
pro_name = ttk.Label(lblframe, text="Product Name:")
name_entry = ttk.Entry(lblframe, textvariable=PRO_NAME, width=25)
pro_price = ttk.Label(lblframe, text="Selling Price:")
price_entry = ttk.Entry(lblframe, textvariable=PRO_PRICE, width=25)
pro_quant = ttk.Label(lblframe, text="Quantity:")
quant_entry = ttk.Entry(lblframe, textvariable=PRO_QUANT, width=25)
insert_button = ttk.Button(lblframe, text="Insert", command=insert_info)
show_button = ttk.Button(lblframe, text="Show", command=show_info)

style = ttk.Style()
style.configure('TButton',background='#3498db')

pro_id.grid(column=0,row=0,sticky='W')
id_entry.grid(column=0,row=1,sticky='W')
pro_name.grid(column=0,row=2,sticky='W')
name_entry.grid(column=0,row=3,sticky='W')
pro_price.grid(column=0,row=4,sticky='W')
price_entry.grid(column=0,row=5,sticky='W')
pro_quant.grid(column=0,row=6,sticky='W')
quant_entry.grid(column=0,row=7,sticky='W')
insert_button.grid(column=0,row=8,sticky='W',pady=7)
show_button.grid(column=0,row=8,sticky='E')

tab_control.pack(expand=1,fill="both")
my_frame.pack()

root.mainloop()
```
