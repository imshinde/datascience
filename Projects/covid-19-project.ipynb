# Covid - 19 India Data Analysis

```python
df_covid = pd.read_csv('covid_19_india.csv')
```

### Data Cleaning

```python
df_covid.drop(columns='Sno', inplace=True)
```

```python
df_covid.drop(columns=['ConfirmedIndianNational','ConfirmedForeignNational'], inplace=True)
```

```python
df_covid
```

### Dates into Datetime

```python
df_covid['Date'] = pd.to_datetime(df_covid['Date'])
```

```python
df_covid['Date']
```

### Year, Month & Day Columns

```python
df_covid['Year'] = df_covid['Date'].dt.year
```

```python
df_covid['Month'] = df_covid['Date'].dt.month
```

```python
df_covid['Day'] = df_covid['Date'].dt.day
```

```python
df_covid
```

## Total Cases by State/UnionTerritory

```python
state_df = df_covid.groupby('State/UnionTerritory')[['Cured', 'Deaths', 'Confirmed']].max().sort_values(by='Confirmed', ascending=False).reset_index()
```

```python
state_df
```

## Covid-19 Cases in India

```python
total_cases = df_covid.groupby('State/UnionTerritory')['Confirmed'].max().sum()
```

```python
total_cases
```

```python
total_cured = df_covid.groupby('State/UnionTerritory')['Cured'].max().sum()
```

```python
total_cured
```

```python
total_deaths = df_covid.groupby('State/UnionTerritory')['Deaths'].max().sum()
```

```python
total_deaths
```

```python
active_cases = total_cases - (total_cured + total_deaths)
```

```python
active_cases
```

```python
print(f"Total Confirmed Cases in India: {total_cases}")
print(f"Total Cured Cases in India: {total_cured}")
print(f"Total Deaths in India: {total_deaths}")
print(f"Total Active Cases in India: {active_cases}")
```

### Cases Ratings

```python
recovery_rate = (total_cured / total_cases) * 100
```

```python
death_rate = (total_deaths / total_cases) * 100
```

```python
active_rate = (active_cases / total_cases) * 100
```

```python
print('Recovery rate is {:.2f} %'.format(recovery_rate))
print('Death rate is {:.2f} %'.format(death_rate))
print('Active rate is {:.2f} %'.format(active_rate))
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df_covid)
sns.set_context("notebook")
```

### State wiese covid-19 cases

```python
fig = plt.figure(figsize=(16,8))
sns.barplot(x='State/UnionTerritory', y='Confirmed', data=state_df, color='b')
sns.set_style("dark")
plt.title("State Wise Confirmed Cases")
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.barplot(x='State/UnionTerritory', y='Deaths', data=state_df, color='r')
sns.set_style("whitegrid")
plt.title("State Wise Death Cases")
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.barplot(x='State/UnionTerritory', y='Cured', data=state_df, color='g')
plt.title("State Wise Cured Cases")
plt.xticks(rotation=90)
plt.show()
```

### Scatter Plot Confirmed Vs Cured

```python
fig = plt.figure(figsize=(16,8))
sns.scatterplot(x='Confirmed', y='Cured', data=df_covid)
plt.title("Confirmed Vs Cured Cases")
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.regplot(x='Confirmed', y='Cured', data=df_covid)
plt.title("Regression Line Confirmed Vs Cured Cases")
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.scatterplot(x='Cured', y='Deaths', data=df_covid)
sns.set_style("whitegrid")
plt.title("Cured Cases Vs Death Cases")
plt.show()
```

### Time Series Daily Cases In India

```python
fig = plt.figure(figsize=(16,8))
sns.lineplot(x='Date',y='Confirmed',data=df_covid.groupby('Date')['Confirmed'].sum().reset_index(), color='b', linewidth=5)
sns.set_style("whitegrid")
plt.title("Daily Confirmed Cases In India")
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.lineplot(x='Date',y='Cured',data=df_covid.groupby('Date')['Cured'].sum().reset_index(), color='g', linewidth=5)
sns.set_style("whitegrid")
plt.title("Daily Cured Cases In India")
plt.show()
```

```python
fig = plt.figure(figsize=(16,8))
sns.lineplot(x='Date',y='Deaths',data=df_covid.groupby('Date')['Deaths'].sum().reset_index(), color='r', linewidth=5)
sns.set_style("whitegrid")
plt.title("Daily Deaths In India")
plt.show()
```

### Cases Rating (%)

```python
fig = plt.figure(figsize=(16,8))
plt.pie(x=[recovery_rate, death_rate, active_rate],
       autopct='%.2f%%',
       radius=1.5)
plt.title("Covid Cases Rating In India")
plt.legend(labels=['Recovery','Death','Active'])
plt.show()
```

### Total Cases In India

```python
fig = plt.figure(figsize=(10,8))
sns.barplot(x=['Confirmed','Cured','Deaths'], y=[total_cases,total_cured,total_deaths])
sns.set_context("notebook")
plt.title("Total Covid Cases In India")
plt.show()
```
